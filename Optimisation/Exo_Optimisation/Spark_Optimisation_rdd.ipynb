```
from pyspark import SparkContext
from pyspark.sql import SparkSession
from pyspark.streaming import StreamingContext
from pyspark.sql import functions as F
from pyspark.sql.functions import *
from pyspark.sql.types import *
import json
import findspark
```

```
findspark.init()
```

```
spark = SparkSession.builder.appName('SparkByExamples.com').getOrCreate()
sc=spark.sparkContext
```

```
import random

data =  [random.randint(0,10) for i in range(1000)]
print(data)
```

```
from sparkmeasure import StageMetrics
stagemetrics = StageMetrics(spark)
```

```
# First example, measure metrics aggregated at stage level
stagemetrics.begin()

rdd = sc.parallelize(data)

stagemetrics.end()
```

```
# This prints a report of aggregated metrics values

stagemetrics.print_report()
```

```
# The metrics are exported into a DataFrame and registered as a Temporary View

spark.sql("show tables").show()
```

```
display(sql("select * from PerfStageMetrics"))
```

```
stagemetrics.runandmeasure(globals(), 'sc.parallelize(data)')
```
